### Box Plot for Distributions

Box plots give us a visualization about how the univariate data points are distributed. 

The code implementation below generates a box plot to visualize the distribution of Sepal Length against the different species of the iris dataset


```r
library(ggplot2)
library(ggthemes)
library(datasets)

#sample iris data 
head(iris, 3)
```

```r
box = ggplot( data = iris, aes(x = Species , 
                               y = Sepal.Length, 
                               color = Species) ) + geom_boxplot() +
ggtitle("Box Plot Distribution of Sepal Length by Species") + theme_economist()
box
```

```r
ggsave('./outputs/ggplot_boxplot.jpg', plot = box, dpi = 1200, width = 7, height = 5, units = "in", device='jpg')
```
