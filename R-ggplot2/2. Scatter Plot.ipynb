### Scatter Plot

Scatterplots are useful in providing a visual representation of the spread and distribution of data across some $x$ and $y$ dimensions. The example below uses the well-known iris dataset in R.

```r
library(ggplot2)
library(datasets)
library(ggthemr)
```

```r
ggthemr('flat')

scatter = ggplot(data = iris, aes( x = Sepal.Length, y = Sepal.Width, color = Species)) + 
        geom_point() + labs( title = "Scatter Plot: Sepal Length vs. Sepal Width by Species" ) 
scatter
```

```r
ggsave('./outputs/ggplot_scatter.jpg', plot = scatter, dpi = 1200, width = 7, height = 5, units = "in", device='jpg')
```
