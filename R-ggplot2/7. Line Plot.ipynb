### Line Plots

There are many variables that change over time. Line plots are useful in visualizing data points across time.

In the example below, I use the airpassengers dataset to produce a time series plot of airline passengers over time. A modification is applied to generate dates (in months) as a column.

```r
library(datasets)
library(ggplot2)
library(ggthemr)
```

```r
# Converting the data to time series 
data = as.data.frame(AirPassengers)
data$months = seq(as.Date("1949-01-01"), by = "month", length.out = nrow(data))

colnames(data) <- c("passengers", "months")
head(data)
```

```r
ggthemr('grape')

base = ggplot(data = data, aes( x = months , y = passengers)) + 
       geom_line() +
       labs( title = "Airline Passengers by Month")
base
```

```r
ggsave('./outputs/ggplot_line.jpg', plot = base, dpi = 1200, width = 7, height = 5, units = "in", device='jpg')
```
